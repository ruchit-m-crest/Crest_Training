## EXAMPLE

```python
 #get title of the books having 3 star ratings
import requests
import bs4
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_num = 23
base_url.format(page_num)
```

```python
res= requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
products = soup.select('.product_pod')
```

```python
example = products[0]
```

```python
example
```

```python
str(example)
```

```python
'star-rating Three' in str(example)
```

```python
#example.select('.star-rating.Three')
```

```python
 example.select('a')[1]['title']
```

```python
three_star_titles = []

for n in range(1,51):

    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)

    soup = bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select('.products_pod')

    for book in books:
        if len(book.select('.star-rating.Three')) !=0:
            book_title = book.select('a')[1]['select']
            three_star_titles.append(book_title)
    
```
